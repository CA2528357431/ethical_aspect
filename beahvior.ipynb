```python



plt.figure(figsize=(6, 6))
plt.imshow(rovit_mat)
plt.title("RoViT's classification behavior")
# plt.colorbar()

# Tick marks for classes
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45, ha="right")
plt.yticks(tick_marks, classes)

# Annotating the heatmap with values
for i in range(rovit_mat.shape[0]):
    for j in range(rovit_mat.shape[1]):
        plt.text(j, i, f"{rovit_mat[i, j]}", ha="center", va="center", color="white" if rovit_mat[i, j] > rovit_mat.max()/2 else "black")

plt.ylabel("True Class")
plt.xlabel("Predicted Class")
plt.tight_layout()

plt.savefig("rovit_behavior.png")
plt.show()







```
